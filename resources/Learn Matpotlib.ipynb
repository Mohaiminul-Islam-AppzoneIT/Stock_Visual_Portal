## SubPlots and legends 

```python
days = [1,2,3,4,5]
orginal_price =[10, 50, 32,53,60]
pedicted_price =[12, 55, 39, 55, 89]
```

```python
plt.plot(days, orginal_price,"b-o", label="Orginal Price");
plt.plot(days, pedicted_price,"r-o", label="Orginal Price");
plt.legend();
plt.title("Plot with legends");
plt.xlabel("Weekdays");
plt.ylabel("Prices");
```

### SubPlots

```python
#### First subloplot 

plt.subplot(3,1,1);
plt.plot(days, orginal_price,"b-o", label="Orginal Price");
plt.legend();
plt.xlabel("Weekdays");
plt.ylabel("Prices");
#### Second subloplot 
plt.subplot(3,1,2);
plt.plot(days, pedicted_price,"r-o", label="Orginal Price");
plt.legend();
plt.xlabel("Weekdays");
plt.ylabel("Prices");
#### Third diffrecence subloplot 
price_deffrence = [o - p for o , p in zip(orginal_price,pedicted_price )]
plt.subplot(3,1,3);
plt.bar(days, price_deffrence,color="purple", label="diffrece Price");
plt.legend();
plt.xlabel("Weekdays");
plt.ylabel("deffrence");

plt.tight_layout();
```

### saving figure 

```python
plt.plot(days, orginal_price,"b-o", label="Orginal Price");
plt.plot(days, pedicted_price,"r-o", label="Orginal Price");
plt.legend();
plt.title("Plot with legends");
plt.xlabel("Weekdays");
plt.ylabel("Prices");
plt.savefig("example preditciton test", dpi=300)
```
